## Question 1: Review the loop structures (10 points)
#### Check the example codes of prompting the user for a list of numbers and prints out the maximum and minimum of the numbers at the end when the user enters “done”.
#### Write the program to store the numbers the user enters in a list and use proper Python build-in functions or loops to compute the maximum and minimum numbers after the loop completes.

```python
print("This program calculates and displays the minimum and maximum of a list of user-entered numbers.")

# declare empty list
lst = []
element = ""

while True:
    try:
        while element != "done":
            # take element input from user
            element = input("Please enter a number or type 'done'. ")
            # convert element to integer
            element = float(element)
            # append element to list
            lst.append(element)
            break

    # handle invalid input
    except ValueError:
        if element != "done":
            print("Please enter a number or 'done' to end and print maximum and minimum")

        # ends input loop if 'done' is entered
        if element == "done":
            break

if len(lst) != 0:
    print("Minimum = {}" .format(min(lst)))
    print("Maximum = {}" .format(max(lst)))


else:
    print("\nYou did not enter any numbers.")
    print("Program ended")
```

## Question 2: Strings (8 points)
#### Given the following string, write a Python script to save it in a string variable.
#### str = 'X-DSPAM-Confidence: 0.8475'
#### Use find and string slicing to extract the portion of the string after the colon character and then use the float function to convert the extracted string into a floating point number.

```python
# store original string as variable
original = 'X-DSPAM-Confidence: 0.8475'

# position of the space in the string (returns 19)
space = original.find(" ")

# initiate slice after space
number = slice(space+1, len(original), 1)

# slice string and store number as float
new = float(original[number])

# print number and data type
print("{} {}" .format(new, type(new)))
```

## Question 3: Strings (12 points)
#### Given the string “supercalifragilisticexpialidocious”.
#### 1. What is the length of the string i.e. write the script to find the length? (2 points)
#### 2. Find the sub-string of first 5 positions and the other sub-string with last 5 positions. (4 points)
#### 3. Find how many times “i” occurs in this word. (6 points)

```python
# store the string as a variable
message = "supercalifragilisticexpialidocious"

# print the length of this string
print("Number of characters in the string: {}" .format(len(message)))

# print sub string of first 5 positions
print("The first 5 letters of the string are: {}" .format(message[:5]))

# print sub string of last 5 positions
print("The last 5 letters of the string are: {}" .format(message[-5:]))

# count and print the amount of
print("Number of times the letter 'i' appears in the string: {}" .format(message.count("i")))
```

## Question 4: Strings (5 points)
#### Given the string “Alex, EyeColor:Black, HairColor:black, Major: Math, School:SoIC”. Split the string according to a proper delimiter. Then display separated tokens after this splitting.

```python
# store the original string as a variable
original = "Alex, EyeColor:Black, HairColor:black, Major:Math, School:SoIC"

# split at comma and space
splitString = (original.split(", "))

for i in range(len(splitString)):
    print(splitString[i])
```

```python
# function to create list
def makeList():
    # declare element and list variables
    originalList = []
    element = input("Please enter a list element or '*' to finish. ")
    originalList.append(element)

    # while loop to store elements and append to list and quit on '*'
    while element != "*":
        element = input("Please enter the next list element or '*' to finish. ")
        originalList.append(element)

    # remove the '*' element from the end of the list
    del originalList[-1]

    print("\nYour list has been stored\n")

    return originalList

# menu for user to select operation to be performed on list
def menu(lst):
    # select operation to perform on list
    selection = input("1. Chop: Return the first and last elements of the list."
                      "\n2. Middle: Return all but the first and last elements."
                      "\ntype '*' to quit. ")

    while selection != "*":

        if selection == "1":
            chop(lst)
            break

        elif selection == "2":
            middle(lst)
            break

        else:
            print("\nPlease choose from the following")
            selection = input("1. Return the first and last elements of the list."
                               "\n2. Return all but the first and last elements."
                                "\ntype '*' to quit. ")


# chops first and last element and prints remaining list
def chop(lst):
    newList = []

    # remove first and last elements from a list of 3 or more elements
    if len(lst) >= 3:

        newList.append(lst.pop(0))
        newList.append(lst.pop(-1))

        #print new list
        print("\nYour new list is: ")
        for i in range(len(newList)):
            print(newList[i])

    # return user to makeList function to create a longer list otherwise
    else:
        print("You must enter at least 3 elements to use the chop function")
        # return to makeList() function
        makeList()


# chops first and last element and prints remaining list
def middle(lst):

    # remove first and last elements from a list of 3 or more elements
    if len(lst) >= 3:

        del lst[0]
        del lst[-1]

        #print new list
        print("\nYour new list is: ")
        for i in range(len(lst)):
            print(lst[i])

    # return user to makeList function to create a longer list otherwise
    else:
        print("You must enter at least 3 elements to use the chop function")
        # return to makeList() function
        makeList()


# call makeList function
storedList = makeList()

# call menu
menu(storedList)
print("\nProgram ended.")
```
