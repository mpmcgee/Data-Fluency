# Guessing Game

```python
from random import randint

#generate drandom number between 1 and 9
num = randint(1, 9)

#declare variable to count number of guesses
times = 1

# while loop for input validation
while True:
    try:
        #get guess from user
        guess = input("Guess enter a number from 1 to 9: ")
        
        #if user does not type exit, convert guess to integer
        if guess != "exit":
            guess = int(guess)

    except ValueError:
            #input validation
            print("Please enter a numerical value.")
            continue

    else:
        break

#while loop for guessing
while (guess!= num and guess!="exit"):

    #let user knof if guess is to low and ask for input again
    if (guess < num):
        print("Too low, try again.")
        #add one to guess count
        times = times+1
        
    #break loop if user types exit    
    elif (guess == "exit"):
        break
        
    #let user knof if guess is to high and ask for input again 
    else:
        print("Too high, try again.")
        #add one to guess countS
        times = times+1

    #take new guess
    while True:
        try:
            guess = input("Guess enter a number from 1 to 9: ")
            if guess != "exit":
                #convert guess to integer
                guess = int(guess)
        
        #input validation
        except ValueError:
                print("Please enter a numerical value.")
                continue

        else:
            break

# if guesses the number wins inform them            
if guess == num:
    print("You guessed the number!")
# display guess count    
if times > 1:
    print("It took you {} guesses" .format(times))
else:
    print("You got it on your first guess!")

print("Program ended")
```

# 3. Rock Paper Scissors

```python
import random

#define play function
def play(winner, player_weapon, ai_weapon):
    
    #while loop to determine winner based on weapons chosen
    while winner == "none":
        if player_weapon == 1 and ai_weapon == "paper":
            winner = "ai"

        elif player_weapon == 1 and ai_weapon == "scissors":
            winner = "player"

        elif player_weapon == 2 and ai_weapon == "rock":
            winner = "player"

        elif player_weapon == 2 and ai_weapon == "scissors":
            winner = "ai"

        elif player_weapon == 3 and ai_weapon == "rock":
            winner = "ai"

        elif player_weapon == 3 and ai_weapon == "paper":
            winner = "player"

        else:
            winner = "tie"

        # print who won or if it is a tie
        if winner == "player":
            print("You won!")
        elif winner == "ai":
            print("Computer Wins!")
        elif winner == "tie":
            print("It's a tie")
        
        #call play again function
        play_again()


# define turn function
def turn():
    #declare winner variable
    winner = "none"
    
    #while loop for input validation
    while True:
        try:
            #print options and take input
            print("1. Rock\n2. Paper\n3. Scissors")
            player_weapon = input("Choose your weapon: ")
            if player_weapon != "exit":
                player_weapon = int(player_weapon)
            
            #select random weapon for AI and inform player
            weapon = ["rock", "paper", "scissors"]
            ai_weapon = random.choice(weapon)
            print("Computer chooses {}" .format(ai_weapon))
            
            #call play function
            play(winner, player_weapon, ai_weapon)

        except ValueError:
            print("1. Rock\n2. Paper\n3. Scissors")
            player_weapon = input("Choose your weapon:")
            continue
            play(winner, player_weapon, ai_weapon)

        else:
            break

#define play again function
def play_again():
    #ask player to input y to play again or n to end program
    print("Would you like to play again?")
    again = input("Type 'y' for yes or 'n' for no. ")
    
    #call turn function to restart game on 'y', or end on 'n;'
    if again == "y":
        turn()
    else:
        print("Program ended")

#print welcome and how to exit
print("Welcome to Rock, Paper, Scissors.")
print("Type 'exit' to end program.")

#call turn function to start game
turn()
```

# 4. Write own functions to print sentences

```python
#define list_benefits function
def list_benefits():
    
    #declare 4 string messages
    msg1 = "More organized code"
    msg2 = "More readable code"
    msg3 = "Easier code reuse"
    msg4 = "Allowing programmers to share and connect code together"
    
    #store list messages in a list
    msg_list = [msg1, msg2, msg3, msg4]
    
    #return the list
    return msg_list

#define build_sentence function
def build_sentence(benefit):
    
    #build sentences out of the benefits
    sentence = "%s is a benefit of functions." % benefit
    
    #return sentences
    return sentence

#define function to print sentences
def name_the_benefits_of_functions():
    list_of_benefits = list_benefits()
    
    #for loop to print sentences
    for benefit in list_of_benefits:
        print(build_sentence(benefit))

#call function to print sentences
name_the_benefits_of_functions()
```
