## 1. Rewrite your pay computation with 1.5 times for overtime and create a function called computePay which takes two parameters (hours and rate).

```python
# prompt input for number of hours worked
hours = input("Enter Hours: ")

# prompt input for rate
rate = input("Enter Rate: ")


def computePay(hours, rate):
    # try except statement to exit program if non-numerical data is entered
    try:
        # convert input to floats
        hours = float(hours)
        rate = float(rate)

        # calculate total pay
        if hours <= 40:
            pay = hours * rate
        else:
            pay = (rate * 40) + (rate*(1.5 * (hours - 40)))

        # print pay with two significant figures
        print("Pay: ${:.2f}".format(pay))

    # print error if non-numerical data was entered
    except:
        print("Please enter numerical data")

computePay(hours, rate)
```

## 2. Rewrite the Question 3 as a function that is called computeGrade that takes a score as its parameter and returns a grade as a string.

```python
# Converts a score from 0.0 to 1.1 into corresponding letter grade

# loop for input validation
while True:

    # take input
    try:
        # get numerical score input and convert to float
        score = float(input("Enter Score: "))

        # while loop to validate that input is in acceptable range
        while score < 0 or score > 1:
            # get score input and convert to float
            print("Please enter a numerical value between 0.0 and 1.0.")
            score = float(input("Enter Score: "))

    # if values is non-numerical, print message and repeat loop
    except ValueError:
        print("Invalid input.")
        print("Please enter a numerical value.")
        continue

    else:
        break


# print letter grade equivalent of to numerical value entered
def computeGrade(num):
    if num >= 0.9:
        grade = "A"
    elif num >= 0.8:
        grade = "B"
    elif num >= 0.7:
        grade = "C"
    elif num >= 0.6:
        grade = "D"
    elif num < 0.6:
        grade = "F"
    else:
        print("Error")

    print(grade)


computeGrade(score)
```

## 3. Write a program that contains a function with name CelToFah. The function has a parameter that is the value of Celsius degree, and the return value is the converted Fahrenheit degree. (Find the conversion formula).

```python
# temperature calculator

# store degree sign as variable
d = u"\u00b0"

# print welcome
print("Welcome to the temerpature converter.")


# function to convert farenheit to celsius
def FarToCel(far):
    temp = (far-32)*(5/9)
    print("{:.1f}{}F is equal to {:.1f}{}C".format(far, d, temp, d))

# function to convert celsius to farenheit
def CelToFar(cel):
    temp = (cel*(9/5)) + 32
    print("{:.1f}{}C is equal to {:.1f}{}F".format(cel, d, temp, d))

# menu to select direction of conversion
def menu():
    # print options
    print("1. {}F to {}C\n2. {}C to {}F ".format(d, d, d, d))

    # prompt user for selection
    selection = (input("What would you like to convert? "))

    # get temperature from user and call selected function
    if selection == "1":
        temp = float(input("Enter the temperature you would like to convert into Celsius: "))
        FarToCel(temp)

    elif selection == "2":
        temp = float(input("Enter the temperature you would like to convert into Farenheit: "))
        CelToFar(temp)

    # input validation
    if selection != "1" and selection != "2":
        print("Please enter '1' for option 1 or '2' for option 2.")
        menu()

# call menu function
menu()
```

## 4. Write a program which repeatedly reads numbers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect the mistake by using try and except structure, and print an error message and skip to the next number.

```python
#read and display numbers entered from a user until they enter done

# function to enter numbers into a list
def readNum():

    #declare variables
    i = 0
    numList = []
    num = ""

    # loop for input validation
    while True:
        try:
            # converts numbers entered into integers and appends them to the list
            while num != "done":
                num = input("Enter a Number: ")
                num = float(num)
                numList.append(num)
                i = i + 1
                break;

        # handle invalid input
        except ValueError:
            if num != "done":
                print("Please enter a number or 'done' to end input and print all numbers entered")

        # ends input loop if 'done' is entered
        if num == "done":
            break

    # calls function to print list
    printNum(numList)


# prints list
def printNum(numList):

    print("\nHere are the numbers you entered:")

    for i in range(len(numList)):
        print(numList[i-1])


# call function to enter numbers into a list
readNum()
```

## 6. Write a Python program that prints out a table with Fahrenheit degrees 0, 10, 20, . . . , 100 in the first column and the corresponding Celsius degrees in the second column.

```python
# print a table showing degrees F 0, 10, 20..., 100 in one column and conversions in another

# import modules for matrix manipulation and dataframe presentation
from pandas import *
from numpy import *

# store degree symbol as d
d = u"\u00b0"

#function to convert 2d array to dataframe and print as table
def printTable(table):

    df = DataFrame(table, columns=[d + "F", d + "C"])
    print (df.to_string(index=False))

#function to set column 1 of 2d array equal to degrees F and 2 as C
def setValues():
    temps = []

    for i in range(11):
        temps.append([i*10, round((((i*10)-32)*(5/9)), 1)])

    #call function to print table
    printTable(temps)



setValues()
```

## 7. Write a program that generates all odd numbers from 1 to n. Set n in the beginning of the program and use a while loop to compute the numbers. (Make sure that if n is an even number, the largest generated odd number is n-1.)

```python
# prints all odd numbers from 1 to n

# prompt user for n
n = int(input("Input a number: "))

# declare lower bound as 1
x = 1


# loop to print numbers while x is less than n
while x < n:

    # if x/2 has a remainder, the number is odd and is printed
    if x % 2 !=0:
        print(x)
    # increment x
    x = x+1

```
